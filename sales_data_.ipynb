**Research Questions: 1**

Q1 Which customers demonstrate the highest levels of loyalty based on Recency, Frequency, and Monetary (RFM) analysis, and to what extent do they contribute to overall business revenue?

**Domain Knowledge Integration:**

* The loyalty of customers is an important element in ensuring that the profitability of businesses remains at an optimal level in the long run. RFM analysis is one of the common approaches employed to segment customers. The three elements considered in RFM include recency, frequency, and monetary value.

```python
# Convert Order Date into datetime format
products['Order Date'] = pd.to_datetime(products['Order Date'], dayfirst=True)
```

```python
latest_date = products['Order Date'].max()
rfm = products.groupby('Customer ID').agg({
    'Order Date': lambda date: (latest_date - date.max()).days
})
rfm.rename(columns={'Order Date': 'Recency'}, inplace=True)
print(rfm.head())
```

```python
rfm['Frequency'] = products.groupby('Customer ID')['Order ID'].nunique()
print(rfm.head())
```

```python
rfm['Monetary'] = products.groupby('Customer ID')['Sales'].sum()
print(rfm.head())
```

```python
# Calculate RFM scores using quintiles
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1],duplicates='drop') # Lower recency is better
rfm['F_score'] = pd.qcut(rfm['Frequency'], 5, labels=[1, 2, 3, 4, 5],duplicates='drop') # Higher frequency is better
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5],duplicates='drop') # Higher monetary is better

# Combine scores to create an RFM score
rfm['RFM_Score'] = rfm['R_score'].astype(int) + rfm['F_score'].astype(int) + rfm['M_score'].astype(int)

print(rfm.head())
```

```python
# Identify Top Loyal Customers (with top rfm score)
top_loyal_customers = rfm[rfm['RFM_Score'] >= rfm['RFM_Score'].quantile(0.75)] # Top 25% by RFM Score

# Sort by RFM score to see the absolute top customers
top_loyal_customers = top_loyal_customers.sort_values(by='RFM_Score', ascending=False)

print("\nTop Loyal Customers (first 10):\n")
print(top_loyal_customers.head(10))
```

```python
# Calculate the total sales by loyal customers
total_sales_loyal_customers = top_loyal_customers['Monetary'].sum()

# Calculate overall total sales
overall_total_sales = rfm['Monetary'].sum()

# Calculate their contribution to overall revenue
revenue_contribution = (total_sales_loyal_customers / overall_total_sales) * 100

print(f"\nTotal sales from top loyal customers: ${total_sales_loyal_customers:,.2f}")
print(f"Overall total sales: ${overall_total_sales:,.2f}")
print(f"Revenue contribution of top loyal customers: {revenue_contribution:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.scatterplot(data=rfm, x='Frequency', y='Monetary')

plt.title("Customer Loyalty Analysis: Frequency vs Monetary Value")
plt.xlabel("Purchase Frequency")
plt.ylabel("Total Customer Spend (Monetary)")
plt.show()
```

**Research Question:2**

Which product categories generate the highest revenue and profitability?

**Domain Knowledge Integration:**

* The need for knowledge of the profitability and sales of categories is necessary in retail and logistics management. Even high sales do not necessarily translate into profits because the margin of the products sold can be low due to other factors such as discounting.

To determine which product categories generate the highest revenue and profitability, we'll analyze the total sales and profit for each category.

```python
# Calculate revenue by category
revenue_by_category = products.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("\nRevenue by Category:\n")
print(revenue_by_category)
```

```python
# Calculate profit by category
profit_by_category = products.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print("\nProfit by Category:\n")
print(profit_by_category)
```

```python
#Visualizing Revenue and Profit by Category
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(x=revenue_by_category.index, y=revenue_by_category.values, ax=axes[0], palette='viridis')
axes[0].set_title('Total Revenue by Category')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Total Revenue')
axes[0].ticklabel_format(style='plain', axis='y') # Prevent scientific notation

sns.barplot(x=profit_by_category.index, y=profit_by_category.values, ax=axes[1], palette='plasma')
axes[1].set_title('Total Profit by Category')
axes[1].set_xlabel('Category')
axes[1].set_ylabel('Total Profit')
axes[1].ticklabel_format(style='plain', axis='y') # Prevent scientific notation

plt.tight_layout()
plt.show()
```

Based on analysis:

* **Highest Revenue:** The Technology category generates the highest revenue, followed closely by Office Supplies and then Furniture.

* **Highest Profitability:** Similarly, Technology leads in terms of total profit, with Office Supplies being the second most profitable. The Furniture category is significantly less profitable, indicating potential issues with pricing, costs, or discount strategies within this category.

**Research Question:3**

Are there seasonal or monthly trends in sales performance, and how do these trends affect customer purchasing behaviour and business planning?

**Domin Knowledge Integration**

It is very important to use time series analysis in the operation process of a retail business due to the fact that the consumer’s needs keep changing from month to month and even seasonally. Seasonality allows for better predictions, inventory management, staffing, and promotional campaign scheduling.

```python

# Extract Month and Year
products['Month'] = products['Order Date'].dt.month
products['Year'] = products['Order Date'].dt.year
products['Month_Name'] = products['Order Date'].dt.month_name()

#Monthly Sales Analysis
monthly_sales = products.groupby('Month_Name')['Sales'].sum().reset_index()

print("Monthly Sales Summary:")
print(monthly_sales)

#Arrange months in correct calendar order
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

monthly_sales['Month_Name'] = pd.Categorical(
    monthly_sales['Month_Name'],
    categories=month_order,
    ordered=True
)

monthly_sales = monthly_sales.sort_values('Month_Name')

# Visualisation — Monthly Sales Trend

plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales, x='Month_Name', y='Sales', marker='o')

plt.title("Monthly Sales Trend Analysis")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

# Identify Peak Sales Month

peak_month = monthly_sales.loc[monthly_sales['Sales'].idxmax()]

print("\nPeak Sales Month:")
print(peak_month)
```
